## Meta-Learning Algorithms

##### Reading dataset

```python
import pandas as pd
df = pd.read_csv('weatherDef.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
X = df[df.columns[:-1]]
y = df[df.columns[-1]]
X.head()
```

##### Train / Test partitions

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0, stratify = X['Loc'])
```

#### Scaling

The only algorithm here that might need scaled data is KNN, forming part of the voting algorithm. We applied MinMax and Standard scalers to our dataset to check if we get an improvement on the KNN, but we did not. So, from now on, we will only use the dataset with the original non-scaled data.

Bagging, Random Forest and Adaboost algorithms do not need any kind of scaling.

#### Majority Voting

We will use Naive Bayes, KNN and Decision Trees on our majority voting algorithm

##### Naive Bayes

clf1 = naive bayes classifier

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score


clf1 = GaussianNB().fit(X_train, y_train)
```

Individual results:

```python
from sklearn.metrics import  ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from yellowbrick.classifier.rocauc import roc_auc

print("ACCURACY:",accuracy_score(clf1.predict(X_test), y_test))
print("CLASSIFICATION REPORT:\n",classification_report(clf1.predict(X_test), y_test));
print("CONFUSION MATRIX AND ROC CURVE:")
plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(clf1, X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
plt.figure(figsize=(8,8));
roc_auc(clf1, X_train, y_train, X_test, y_test, binary=True, classes=['No', 'Yes']);
```

##### KNN


Finding best parameters using GridSearch

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV


params = {'n_neighbors':list(range(1,50,2)), 'weights':('distance','uniform')}
knc = KNeighborsClassifier()


clf = GridSearchCV(knc, param_grid=params,cv=5,n_jobs=-1) 
clf.fit(X, y)


print("Best Params for Knn:",clf.best_params_)

parval=clf.best_params_
```

clf2 = knn classifier

```python
clf2 = KNeighborsClassifier(n_neighbors=parval['n_neighbors'],weights=parval['weights'])
clf2.fit(X_train,y_train);
```

Individual results:

```python

print("ACCURACY:",accuracy_score(clf2.predict(X_test), y_test))
print("CLASSIFICATION REPORT:\n",classification_report(clf2.predict(X_test), y_test));
print("CONFUSION MATRIX AND ROC CURVE:")

plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(clf2, X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
plt.figure(figsize=(8,8));
roc_auc(clf2, X_train, y_train, X_test, y_test, binary=True, classes=['No', 'Yes']);
```

##### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier

clf3 = DecisionTreeClassifier(criterion='entropy') 
clf3.fit(X_train,y_train);
```

```python
print("ACCURACY:",accuracy_score(clf1.predict(X_test), y_test))
print("CLASSIFICATION REPORT:\n",classification_report(clf1.predict(X_test), y_test));
print("CONFUSION MATRIX AND ROC CURVE:")
plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(clf3, X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
plt.figure(figsize=(8,8));
roc_auc(clf3, X_train, y_train, X_test, y_test, binary=True, classes=['No', 'Yes']);
```

##### Majority Voting algorithm

Execute algorithm with previous classifiers:

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
from sklearn.ensemble import VotingClassifier

mvclf = VotingClassifier(estimators=[('nb', clf1), ('knn', clf2), ('dt', clf3)], voting='hard')
mvclf.fit(X_train,y_train);
```

Voting results:

```python
print("ACCURACY:",accuracy_score(mvclf.predict(X_test), y_test))
print("CLASSIFICATION REPORT:\n",classification_report(mvclf.predict(X_test), y_test));
plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(mvclf, X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
```

#### Bagging

Next we are going to use bagging, using a decision tree classifier as the base estimator. We will try different parameter configurations, like the number of estimators and the maximum depth of the decision tree used as the base estimator.

```python
from sklearn.ensemble import BaggingClassifier

lb=[]
for nest in [100,200,400]:
    for md in [5, 20, 25, 30]:
        bclf = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth=md),n_estimators=nest)
        bclf.fit(X_train,y_train)
        score = accuracy_score(bclf.predict(X_test), y_test)
        print("Accuracy: %0.3f [%s,%s]" % (score, nest, md))
        lb.append((score,[nest,md],bclf))
```

```python
sc = [a for (a,b,c) in lb]
sci = sc.index(max(sc))
best_bclf = lb[sci][2]
print("ACCURACY:",accuracy_score(best_bclf.predict(X_test), y_test))
print("BEST PARAMS:", lb[sci][1])
print("CLASSIFICATION REPORT:\n",classification_report(best_bclf.predict(X_test), y_test));
print(classification_report(best_bclf.predict(X_test), y_test));
plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(best_bclf, X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
plt.figure(figsize=(8,8));
roc_auc(best_bclf, X_train, y_train, X_test, y_test, binary=True, classes=['No', 'Yes']);
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

lrf=[]
best_rfclf=None
for nest in [300,400]:
    for md in [10,15,20,25,30,40]:
        for mf in [10,20,25,30,40]:
            rfclf = RandomForestClassifier(max_depth=md, n_estimators=nest, max_features=mf)
            rfclf.fit(X_train, y_train)
            score = accuracy_score(rfclf.predict(X_test), y_test)
            print("Accuracy: %0.3f [%s,%s,%s]" % (score, nest, md, mf))
            lrf.append((score,[nest,md,mf],rfclf))
```

We can se the best result is 0.786 obtained with nest=300, md=40 and mf=10.

```python
sc = [a for (a,b,c) in lrf]
sci = sc.index(max(sc))
best_rfclf = lrf[sci][2]
print("ACCURACY:",accuracy_score(best_rfclf.predict(X_test), y_test))
print("BEST PARAMS:", lrf[sci][1])
print("CLASSIFICATION REPORT:\n",classification_report(best_rfclf.predict(X_test), y_test));
plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(best_rfclf, X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
plt.figure(figsize=(8,8));
roc_auc(best_rfclf, X_train, y_train, X_test, y_test, binary=True, classes=['No', 'Yes']);
```

#### Adaboost


Adaboost starting empty

```python
from sklearn.ensemble import AdaBoostClassifier

lboo=[-1]
best_abclf=None
for nest in [1,2,5,10,20,50,100,200,400]:
    abclf = AdaBoostClassifier(n_estimators=nest)
    abclf.fit(X_train,y_train)
    score = accuracy_score(abclf.predict(X_test), y_test)
    if lboo[-1] < score:
        lboo[-1] = score
        best_abclf = abclf
    print("Accuracy: %0.3f [%s]" % (score, nest))
    lboo.append(score)
```

The best results we get uses 400 estimators and gets an accuracy of 0.773.

```python
print(classification_report(abclf.predict(X_test), y_test));
plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(abclf, X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
plt.figure(figsize=(8,8));
roc_auc(abclf, X_train, y_train, X_test, y_test, binary=True, classes=['No', 'Yes']);
```

Adaboost starting with decision trees

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

lboodt=[]
best_ab2clf = None
for nest in [200,400]:
    for md in [5,10,15,20,30,40]:
        ab2clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=md),n_estimators=nest)
        ab2clf.fit(X_train,y_train)
        score = accuracy_score(ab2clf.predict(X_test), y_test)
        print("Accuracy: %0.3f [%s,%s]" % (score, nest, md))
        lboodt.append((score,[nest,md],ab2clf))
```

Our besr result is 0.789. We obtain it with 200 estimators using a decision tree with max depth of 15 of base estimator.

```python
print(classification_report(lboodt[2][2].predict(X_test), y_test));
plt.figure(figsize=(8,8));
ConfusionMatrixDisplay.from_estimator(lboodt[2][2], X_test, y_test, display_labels=['No', 'Yes'], ax=plt.subplot());
plt.figure(figsize=(8,8));
roc_auc(lboodt[2][2], X_train, y_train, X_test, y_test, binary=True, classes=['No', 'Yes']);
```
